```python
import numpy as np
import math
import time
import random as rn
import copy
import matplotlib.pyplot as plt

class Creatures:
    
    def __init__(self, x, y, speed, visibility, satiety):
        self.x = x
        self.y = y
        self.speed = speed
        self.visibility = visibility
        self.satiety = satiety
    
    def moove(self, x, y): # процесс движения
        if self.x < x:#проверка на направление X
            self.x += self.speed
        elif math.fabs(self.x - x) < self.speed:
            self.x = x
        else:
            self.x -= self.speed
            
        if self.y < y:#проверка на направление y
            self.y += self.speed
        elif math.fabs(self.y - y) < self.speed:
            self.y = y
        else:
            self.y -= self.speed
            
    def eating(self): # получение энергии
        self.satiety += 10
        
    def expenditure(self):#трата энергии
        self.satiety -= 10
    
    def reproduction(self):#размножение
        self.satiety -= 10
        speed = self.speed + rn.randrange(-1,2,1)
        visibility = self.visibility + rn.randrange(-1,2,1)
        return Creatures(self.x, self.y, speed, visibility, 0)
    
    def print(self):
        print('x =', self.x, 'y =', self.y,'скорость =', self.speed,'обзор =', self.visibility, 'сытость =', self.satiety)
        
x = np.arange(0, 100, 1) #создаем 2d пространтсво
y = np.arange(0, 100, 1)
creature = [Creatures(0, len(y)/2, 1, 10, 0)] #список существ с параметрами представлеными списком
allspeed = []
atr1 = []
atr2 = []
allvisibility = []
allpopulation = 0 
average_speed = []

Generetion = 10 #количество поколений
for i in range(Generetion):
    i += 1
    print("Поколение", i)
    
    food = [[rn.randrange(100),rn.randrange(100)] for x in range(5)] #список еды с рандомными координатами
    point_objectiv = [0, 0] 
    while len(food) != 0 :
        food_distand_for_all = [[math.hypot(food[i][0] - creature[j].x,
                               food[i][1] - creature[j].y)for i in range(len(food))]for j in range(len(creature))]
        # нахождение дистанции от существа до всей еды
        for i in range(len(creature)):
            if min(food_distand_for_all[i]) < creature[i].visibility:
                point_objectiv = [0,0] #при нахождение еды точка к которой двигался обнуляется
                point_food = food[np.argmin(food_distand_for_all[i])]
                creature[i].moove(point_food [0], point_food[1])
                if creature[i].x == point_food [0] and creature[i].y == point_food[1]:#проверка на нахождение корма
                    del food[np.argmin(food_distand_for_all[i])]
                    creature[i].eating ()
                    food_distand_for_all = [[math.hypot(food[i][0] - creature[j].x,
                                           food[i][1] - creature[j].y)for i in range(len(food))]for j in range(len(creature))]
                    #обновляем массив дистанции для предотвращения ошибки несоответсвия размеров массива еды и дистанции в цикле
                    break #без возврата в начало for (перехода на следующй i) возникает ошибка,
                          #если следующие существа доходят до уже удаленной еды
            else:#если еду не видит
                if point_objectiv[0] == 0 and point_objectiv[1] == 0 : #назначаем случайную точку
                    point_objectiv[0] =(creature[i].x + 10*rn.randrange(-creature[i].visibility, creature[i].visibility, 1))%100
                    #берем остаток от деления чтоб объект оставался в области карты
                    point_objectiv[1] =(creature[i].y + 10*rn.randrange(-creature[i].visibility, creature[i].visibility, 1))%100
                else:
                    creature[i].moove(point_objectiv[0], point_objectiv[1])
                    if creature[i].x == point_objectiv[0] and creature[i].y == point_objectiv[1]:
                        point_objectiv[0] = 0
                        point_objectiv[1] = 0
       
    #создадим кортеж этого поколения
    Gen = tuple(creature)
    dethnote = []
    
    for cr in creature:cr.print()
    
    for i in range(len(Gen)):
        cost_vital = Gen[i].satiety//(10)
        if cost_vital >= 1 :
            cost_vital -= 1
            creature[i].expenditure()
            for j in range(cost_vital):            
                creature.insert(len(creature),creature[i].reproduction())
        else:
            dethnote += [i] #создает "расстрельный список"
    dethnote_exe = dethnote.copy()
    dethnote_exe.reverse()
    for i in dethnote_exe:
        dethnote.remove(i)
        creature.pop(i)
    dethnote_exe.clear()
    
    for i in range(len(creature)): 
        allspeed.append(creature[i].speed)
        allvisibility.append(creature[i].visibility)
        allpopulation += 1
    
    #for i in range(len(creature)):creature[i].print()
    print("После еды")
    for cr in creature:cr.print()
    
    fig, ax = plt.subplots(figsize=(6, 3))
    
    ax.set_title('Гистограмма всех существ')
    ax.set_ylabel('Количество особей')
    ax.set_xlabel('Скорость')

    for i in range(min(allspeed), max(allspeed) + 1, 1):
        atr1.append(allspeed.count(i))
        atr2.append(i)

    ax.bar(atr2, atr1)

    ax.set_facecolor('seashell')
    fig.set_facecolor('floralwhite')

    plt.show()
    
    atr3 = []
    atr4 = []
    speedGen = []
    for j in range(len(creature)):
        speedGen.append(creature[j].speed)
        
    fig, ax = plt.subplots(figsize=(6, 3))
    
    ax.set_title('Гистограмма этого поколения')
    ax.set_ylabel('Количество особей')
    ax.set_xlabel('Скорость')

    for i in range(min(speedGen), max(speedGen) + 1, 1):
        atr4.append(speedGen.count(i))
        atr3.append(i)

    ax.bar(atr3, atr4)

    ax.set_facecolor('seashell')
    fig.set_facecolor('floralwhite')

    plt.show()
    
    time.sleep (0)
    
    generetion_speed = [creature[i].speed for i in range(len(creature))]
    average_speed.append(np.mean(generetion_speed))
    
fig, ax = plt.subplots(figsize=(6, 3))

ax.set_title('Зависимость средней скорости')
ax.set_ylabel('Средняя скорость')
ax.set_xlabel('Поколение')

plt.plot(range(len(average_speed)), average_speed )

plt.show() 

print("end sim")    
    
```
