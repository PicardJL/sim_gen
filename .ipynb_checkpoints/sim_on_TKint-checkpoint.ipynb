```python
from tkinter import *
import math
import time
import random as rn
import numpy as np
import matplotlib.pyplot as plt
import sys
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import matplotlib

class Creatures:
    
    def __init__(self, x, y, speed, visibility, satiety, canvas):
        self.x = x
        self.y = y
        self.speed = speed
        self.visibility = visibility
        self.satiety = satiety
        self.canvas = canvas
        self.id = canvas.create_oval(0,0, 10, 10, fill='Green')
        self.canvas.move(self.id, self.x, self.y)
    
    def moove(self, x, y): # процесс движения
        diff_x = self.x#запоминаем значение до движения по х
        if self.x < x:#проверка на направление X
            self.x += self.speed
        elif math.fabs(self.x - x) < self.speed:
            self.x = x
        else:
            self.x -= self.speed
        diff_x -= self.x #находим разницу для смены координат в фун-ии move 
        diff_y = self.y
        if self.y < y:#проверка на направление y
            self.y += self.speed
        elif math.fabs(self.y - y) < self.speed:
            self.y = y
        else:
            self.y -= self.speed
        diff_y -= self.y
        self.canvas.move(self.id, -diff_x, -diff_y)#разница берем отрицательную             
            
    def eating(self): # получение энергии
        self.satiety += 10
        
    def expenditure(self):#трата энергии
        self.satiety -= 10
    
    def reproduction(self):#размножение
        self.satiety -= 10
        speed = self.speed + rn.randrange(-1,2,1)
        visibility = self.visibility + rn.randrange((-self.visibility*0.1)//1,(self.visibility*0.1)//1,1)
        return Creatures(self.x, self.y, speed, visibility, 0, canvas)
    
    def print(self):
        print('x =', self.x, 'y =', self.y,'скорость =', self.speed,'обзор =', self.visibility, 'сытость =', self.satiety)
        
    def shift(self):#перестановка существ перед новым циклом
        self.canvas.move(self.id, -self.x, -self.y)#передвижение в координату 0 0
        if rn.choice((0, 1)) == 0:#случайный выбор стороны 
            self.x = 0
            self.y = rn.randint(0, 500)
        else:
            self.y = 0
            self.x = rn.randint(0, 500)
        self.canvas.move(self.id, self.x, self.y)#перемещение на новые случайные координаты
        
    def clean(self):
        self.canvas.delete(self.id)
        
        
class Food:
    
    def __init__(self, x, y, canvas):
        self.x = rn.randrange(width)
        self.y = rn.randrange(height)
        self.canvas = canvas
        self.id = canvas.create_polygon(0, 0, 10, 0, 5, 5, fill = 'Red')
        self.canvas.move(self.id, self.x, self.y)
        
    def clean(self):
        self.canvas.delete(self.id)
                   

def start():#основной цикл программы добавлен в функцию кнопки
    creature = [Creatures(0, height/2, 1, height*0.15, 0, canvas)] #список существ с параметрами представлеными списком
    allspeed = []
    atr1 = []
    atr2 = []
    allvisibility = []
    allpopulation = 0 
    average_speed = []
    fig2, ax2 = plt.subplots(figsize=(6, 3))

    Generetion = entry_1.get()  #количество поколений 
    for i in range(int(Generetion)):
        i += 1
        print("Поколение", i)

        food = [Food(0, 0, canvas) for x in range(int(entry_2.get()))] #список еды с рандомными координатами
        point_objectiv = [[0, 0] for i in range(len(creature))] 
        while len(food) != 0 :
            #time.sleep(0.0005)#"скорость" работы основного цикла
            
            tk.update_idletasks()#обновление "холста"
            tk.update()
            
            food_distand_for_all = [[math.hypot(food[i].x - creature[j].x,
                                   food[i].y - creature[j].y)for i in range(len(food))]for j in range(len(creature))]
            # нахождение дистанции от существа до всей еды
            for i in range(len(creature)):
                if min(food_distand_for_all[i]) < creature[i].visibility:
                    point_objectiv [i] = [0, 0] #при нахождение еды точка к которой двигался обнуляется
                    number_nearest_food = np.argmin(food_distand_for_all[i])
                    creature[i].moove(food[number_nearest_food].x, food[number_nearest_food].y)
                    #проверка на нахождение корма
                    if creature[i].x == food[number_nearest_food].x and creature[i].y == food[number_nearest_food].y:
                        food[number_nearest_food].clean()
                        del food[number_nearest_food]
                        creature[i].eating ()
                        food_distand_for_all = [[math.hypot(food[i].x - creature[j].x,
                                               food[i].y - creature[j].y)for i in range(len(food))]for j in range(len(creature))]
                        #обновляем массив дистанции для предотвращения ошибки несоответсвия размеров массива еды и дистанции в цикле
                        break #без возврата в начало for (перехода на следующй i) возникает ошибка,
                              #если следующие существа доходят до уже удаленной еды
                else:#если еду не видит
                    if point_objectiv[i][0] == 0 and point_objectiv[i][1] == 0 : #назначаем случайную точку
                        point_objectiv[i][0] =(creature[i].x + 10*rn.randrange(-creature[i].visibility, creature[i].visibility, 1))%width
                        #берем остаток от деления чтоб объект оставался в области карты
                        point_objectiv[i][1] =(creature[i].y + 10*rn.randrange(-creature[i].visibility, creature[i].visibility, 1))%height
                    else:
                        creature[i].moove(point_objectiv[i][0], point_objectiv[i][1])
                        if creature[i].x == point_objectiv[i][0] and creature[i].y == point_objectiv[i][1]:
                            point_objectiv[i][0] = 0
                            point_objectiv[i][1] = 0
        #точка не сохраняется(или обнуляется) для разных существ(внутри основного цикла движения)
        #для всей популяции она одинакова, это ошибка
        #для исправления был добавлен список из таких точек,"персонально" для каждого существа
        for cr in creature:cr.shift()#перестановка существ после окончания еды

        #создадим кортеж этого поколения
        Gen = tuple(creature)
        dethnote = []

        for cr in creature:cr.print()

        for i in range(len(Gen)):
            cost_vital = Gen[i].satiety//(10)
            if cost_vital >= 1 :
                cost_vital -= 1
                creature[i].expenditure()
                for j in range(cost_vital):            
                    creature.insert(len(creature),creature[i].reproduction())
            else:
                dethnote += [i] #создает "расстрельный список"
        dethnote_exe = dethnote.copy()
        dethnote_exe.reverse()#чтобы не нарушать порядок в списке,удалять сначала элементы с большим индексом
        for i in dethnote_exe:
            dethnote.remove(i)
            creature[i].clean()
            creature.pop(i)
        dethnote_exe.clear()


        for i in range(len(creature)): 
            allspeed.append(creature[i].speed)
            allvisibility.append(creature[i].visibility)
            allpopulation += 1

        #for i in range(len(creature)):creature[i].print()
        print("После еды")
        for cr in creature:cr.print()

        fig, ax = plt.subplots(figsize=(6, 3))

        ax.set_title('Гистограмма всех существ')
        ax.set_ylabel('Количество особей')
        ax.set_xlabel('Скорость')

        for i in range(min(allspeed), max(allspeed) + 1, 1):
            atr1.append(allspeed.count(i))
            atr2.append(i)

        ax.bar(atr2, atr1)

        ax.set_facecolor('seashell')
        fig.set_facecolor('floralwhite')

        plt.show()

        atr3 = []
        atr4 = []
        speedGen = []
        for j in range(len(creature)):
            speedGen.append(creature[j].speed)

       # fig2, ax2 = plt.subplots(figsize=(6, 3))

        ax2.set_title('Гистограмма этого поколения')
        ax2.set_ylabel('Количество особей')
        ax2.set_xlabel('Скорость')

        for i in range(min(speedGen), max(speedGen) + 1, 1):
            atr4.append(speedGen.count(i))
            atr3.append(i)

        ax2.bar(atr3, atr4)

        ax2.set_facecolor('seashell')
        fig2.set_facecolor('floralwhite')

        plt.show()

        generetion_speed = [creature[i].speed for i in range(len(creature))]
        average_speed.append(np.mean(generetion_speed))

    fig, ax = plt.subplots(figsize=(6, 3))

    ax.set_title('Зависимость средней скорости')
    ax.set_ylabel('Средняя скорость')
    ax.set_xlabel('Поколение')

    plt.plot(range(len(average_speed)), average_speed )

    plt.show() 

    print("end sim")
        
def end():#завершение программы
    tk.destroy()
    #sys.exit(0)

#matplotlib.use('TkAgg')#каждый новый график выводится в отдельном окне
# создаём новый объект — окно с полем. В нашем случае переменная окна называется tk, и мы его сделали из класса Tk() — он есть в графической библиотеке 
tk = Tk()
# делаем заголовок окна
tk.title('Simular')
# запрещаем менять размеры окна, для этого используем свойство resizable 
tk.resizable(0, 0)
#устанавливаем ширину и высоту поля
width = 500
height = 500
canvas = Canvas(tk, width=width, height=height+150, highlightthickness=0)#150 это количество пикселей для виджетов
canvas.create_line(0, 0, 0, width, height, width, height, 0, 0, 0)#граница поля

lbl_1 = Label(tk, text="Поколений")
lbl_1.place(x=0, y=510)
entry_1 = Entry(tk)
entry_1.insert(0, 5)
entry_1.place(x=100,y=510)

lbl_2 = Label(tk, text="Количество еды")
lbl_2.place(x=0, y=535)
entry_2 = Entry(tk)
entry_2.insert(0, 5)
entry_2.place(x=100,y=535)

#lbl_3 = Label(tk, text="Скорость симуляции")
#lbl_3.place(x=230, y=510)
#entry_3 = Entry(tk)
#entry_3.insert(0, 1)
#entry_3.place(x=350,y=510)

btn_1 = Button(tk, text='Старт симуляции', command=start)
btn_1.place(x=0,y=585)

btn_2 = Button(tk, text='Конец симуляции', command=end)
btn_2.place(x=120,y=585)

canvas.pack()
# обновляем окно с холстом
tk.update()
tk.mainloop()
#добавить графики в визуаизацию?
#проверка
```

```python
import tkinter as tk
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import matplotlib.pyplot as plt
import matplotlib
import math

def updateGraph():
    """Example function triggered by Tkinter GUI to change matplotlib graphs."""
    global currentGraph
    # Clear all graphs drawn in figure
    plt.clf()
    y = []
    if currentGraph == "sin":
        for i in x:
            y.append(math.cos(i))
        currentGraph = "cos"
    else:
        for i in x:
            y.append(math.sin(i))
        currentGraph = "sin"
    plt.plot(x,y)
    fig.canvas.draw()

# This defines the Python GUI backend to use for matplotlib
matplotlib.use('TkAgg')

# Initialize an instance of Tk
root = tk.Tk()

# Initialize matplotlib figure for graphing purposes
fig = plt.figure(1)

# Special type of "canvas" to allow for matplotlib graphing
canvas = FigureCanvasTkAgg(fig, master=root)
plot_widget = canvas.get_tk_widget()

# Example data (note: default calculations for angles are in radians)
x = []
for i in range(0, 500):
    x.append(i/10)
y = []
for i in x:
    y.append(math.sin(i))
plt.plot(x, y)

currentGraph = "sin"

# Add the plot to the tkinter widget
plot_widget.grid(row=0, column=0)
# Create a tkinter button at the bottom of the window and link it with the updateGraph function
tk.Button(root,text="Update",command=updateGraph).grid(row=1, column=0)

root.mainloop()
```
